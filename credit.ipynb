ANALYSING Num_of_Loan COLUMN

```python
df["Num_of_Loan"] = pd.to_numeric(df["Num_of_Loan"],errors  = "coerce")
df["Num_of_Loan"] = df["Num_of_Loan"].astype(float)
```

```python
df["Num_of_Loan"].info()
```

MODIFYING THE Num_of_Delayed_Payment COLUMN

```python
df["Num_of_Delayed_Payment"] = pd.to_numeric(df["Num_of_Delayed_Payment"], errors = "coerce")
df["Num_of_Delayed_Payment"] = df["Num_of_Delayed_Payment"].astype(float)
```

```python
df["Num_of_Delayed_Payment"].fillna(df["Num_of_Delayed_Payment"].mean(),inplace = True)
```

MODIFYING Changed_Credit_Limit COLUMN`

```python
df["Changed_Credit_Limit"] = pd.to_numeric(df["Changed_Credit_Limit"], errors = "coerce")
df["Changed_Credit_Limit"]= df["Changed_Credit_Limit"].astype(float)
```

```python
df["Changed_Credit_Limit"].fillna(df["Changed_Credit_Limit"].mean(),inplace = True)
```

MODIFYING THE Num_Credit_Inquiries COLUMN

MODIFYING Credit_Mix COLUMN

MODIFYING Credit_History_Age COLUMN

```python
df["Credit_History_Age"].info()
```

```python
df['Credit_History_Age'] = df['Credit_History_Age'].apply(lambda x: int(x.split()[0]) + int(x.split()[3])/12.0 if isinstance(x, str) else x)
```

COLUMN Outstanding_Debt MODIFICATION

```python
df["Outstanding_Debt"] = pd.to_numeric(df["Outstanding_Debt"], errors = "coerce")
```

REPLACING VARIABLES WITH CATEGORICAL FEATURES INTO NUMERIC

```python
df["Payment_Behaviour"].replace({"Low_spent_Small_value_payments":0,"High_spent_Medium_value_payments":1,
                                "Low_spent_Medium_value_payments":2,"High_spent_Large_value_payments":3,
                                "High_spent_Small_value_payments":4,"Low_spent_Large_value_payments":5,
                                "!@9#%8":6} , inplace = True)
```

REMOVING THE OUTLIERS USING ZSCORES

```python
from scipy.stats import zscore

threshold = 3

# Iterate over each column
for column_name in df.columns:
    # Calculate Z-scores for the current column
    z_scores = zscore(df[column_name])
    
    # Create a boolean mask for outliers in the current column
    outlier_mask = (abs(z_scores) > threshold)
    
    # Remove rows with outliers from the DataFrame for the current column
    df = df[~outlier_mask]
```
